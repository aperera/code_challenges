```python
def calculate_points(cards):
    points = []

    for card in cards:
        # Extracting card data after 'Card X:'
        card_data = card.split(':', 1)[1].strip()
        winning_numbers, your_numbers = card_data.split(" | ")
        winning_numbers = set(map(int, winning_numbers.split()))
        your_numbers = set(map(int, your_numbers.split()))

        matches = winning_numbers.intersection(your_numbers)
        card_points = 0
        if matches:
            # Doubling for each additional match
            card_points = 1 << (len(matches) - 1)

        points.append(card_points)

    return points


def read_cards_from_file(filename):
    with open(filename, 'r') as file:
        cards = [line.strip() for line in file if '|' in line]
    return cards

# Reading cards from the file
cards = read_cards_from_file("puzzle-input.txt")

# Calculating points for each card
points = calculate_points(cards)

# Calculating the total points across all cards
total_points = sum(points)

# Printing results for each card and the total
for i, point in enumerate(points, 1):
    print(f"Card {i}: {point} points")
print(f"Total points: {total_points}")
```

```python
def count_matches(winning_numbers, your_numbers):
    """Count the number of matches between winning numbers and your numbers."""
    winning_numbers = set(map(int, winning_numbers.split()))
    your_numbers = set(map(int, your_numbers.split()))
    return len(winning_numbers.intersection(your_numbers))


def process_cards_with_originals(cards):
    total_won_cards = [1] * len(cards)  # Start with 1 original for each card

    def win_cards(card_index, copies):
        if card_index >= len(cards):
            return

        winning_numbers, your_numbers = cards[card_index].split(" | ")
        matches = count_matches(winning_numbers, your_numbers)

        # For each match, win copies of subsequent cards
        for i in range(1, matches + 1):
            next_card_index = card_index + i
            if next_card_index < len(cards):
                total_won_cards[next_card_index] += copies
                # Recursively win more copies
                win_cards(next_card_index, copies)

    # Start from each card and calculate wins
    for i in range(len(cards)):
        win_cards(i, 1)  # Initial copy of each card

    return total_won_cards


def read_cards_from_file(filename):
    with open(filename, 'r') as file:
        cards = []
        for line in file:
            if '|' in line:
                # Extract only the numbers part after 'Card X:'
                card_data = line.split(':', 1)[1].strip()
                cards.append(card_data)
    return cards


# Reading cards from the file
cards = read_cards_from_file("puzzle-input.txt")

# Process cards including originals and calculate total won cards
total_won_cards_with_originals = process_cards_with_originals(cards)
total_scratchcards = sum(total_won_cards_with_originals)

# Output the total number of scratchcards
print(f"Total scratchcards: {total_scratchcards}")
```
